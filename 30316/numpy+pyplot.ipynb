```python
t=np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)

plt.plot(t,y1,'--r',marker='*')
plt.plot(t,y2,'go--')

plt.grid()
plt.figure(figsize=(12,6))
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'--r',marker='*')

plt.subplot(1,2,2)
plt.plot(t,y2,'go--')
```

### [문제] 

- subplot 을 이용해서 1행 2열 구조2개씩
- 4가지 색상 다르게
- 4가지 스타일 다르게
- 4가지 마커 서로 다르게
- 범례 표시
- x,y 제목

```python
x=np.arange(0,5,1)
x
```

```python
y1=x+1
y1
```

```python
y2=x+2
y2
```

```python
y3=x+3
y3
```

```python
y4=x+4
y4
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'-r',marker='*')
plt.plot(x,y2,'gx--')
plt.legend('xy')
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(1,2,2)
plt.plot(x,y3,'yo-')
plt.plot(x,y4)
plt.legend('XY')
```
